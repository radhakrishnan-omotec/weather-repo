###A. Box plot:

```python
# BOX PLOT PYTHON CODE SAMPLE -BASEL REGION

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read "Output Responses.CSV File"
data =  pd.read_csv("/content/weather-repo/weather_prediction_dataset_25.csv")

data.head(1)

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Set up the figure and axes
plt.figure(figsize=(16, 16))
sns.set(style="whitegrid")

# Create the boxplot using Seaborn
sns.boxplot(data=df.iloc[:, 2:10], palette="viridis")  # Assuming the numerical columns start from index 1 # DEA Correction to the appropriate name


# Customize the plot
plt.title("Data Collected from Region: BASEL") # DEA Correction to the appropriate name
plt.xticks(rotation=90)
plt.xlabel("x Axis")
plt.ylabel("y Axis")

# Show the plot
plt.tight_layout()
plt.show()
```
